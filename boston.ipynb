```python
from plotnine import *
import plotnine

train_scores, test_scores = list(),list()
values = [i for i in [0.1,0.001,0.05]]

for i in values:
    lasso =Lasso(alpha = i)
    lasso.fit(x_train, y_train)
    train_yhat = lasso.predict(x_train)
    train_acc = r2_score(y_train, train_yhat)
    train_scores.append(train_acc)
    test_yhat = lasso.predict(x_test)
    test_acc = r2_score(y_test, test_yhat)
    test_scores.append(test_acc)
    diff = train_acc-test_acc

    print('Alpha_value:%.4f, train_score: %.4f,test_score: %4f, diff:%.4f' % (i, train_acc, test_acc, diff))
plt.plot(values, train_scores, '-o',label='Train')
plt.plot(values, test_scores, '-o',label='Test')
plt.title("Accuracy Curves With Lasso")
plt.xlabel("Alpha Value")
plt.ylabel("Accuracy Score")
plt.legend()
plt.grid()
plt.show()
```
